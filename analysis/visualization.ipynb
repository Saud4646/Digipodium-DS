```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle=''
marker ='s',
markersize=10,
alpha=0.5,
figsize=(10,5),
markerfacecolor='yellow',
markeredgecolor='black')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(15,5),color='green')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='area',figsize=(15,5),color='green')
```

```python
penguins.plot(kind = 'scatter',x= 'bill_length_mm',y='bill_depth_mm',s=100, alpha=.5)
```

subplots

```python
fig,axs= plt.subplots(1,2,figsize=(15,4), constrained_layout=True)
penguins.plot(kind = 'scatter',x= 'bill_length_mm',y='bill_depth_mm',s=100, alpha=.5, ax=axs[0],title='penguin bill data')
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(15,5),color='green',ax=axs[1],title='penguin flipper')
```

```python
penguins['species'].value_counts()
```
